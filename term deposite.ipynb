# Categorized the numeric column for better Key insights

```python
# creating function for age group

def age_grp(x):
    if x >=18 and x<=30:
        return 'Young Adults'
    elif x>30 and x<=45:
        return 'Mid Age Adults'
    elif x>45 and x<=60:
        return 'Veterans'
    else:
        return 'Senior Citizen'
```

```python
# applying function on column 

bank_df['Age_Group'] = bank_df['Age_Group'].apply(age_grp)
```

```python
# checking value counts after categorizing

bank_df['Age_Group'].value_counts()
```

```python
# job has 288 unknown values so will replace it with blue collar

bank_df.Job_type.value_counts()
```

```python
# Replacing null value with mode column

def unknown_bluecollar(x):
    if x == 'unknown':
        return 'blue-collar'
    else:
        return x
```

```python
# applying function on job column

bank_df['Job_type'] = bank_df['Job_type'].apply(unknown_bluecollar)
```

```python
# Checking values after replacing null values

bank_df.Job_type.value_counts()
```

```python
# checking unique value of job column

bank_df.Job_type.unique()
```

```python
# creating function for job group
# considered desk job as white collar job
# considerd field job as blue collar job
# considerd self-employed as Entrepreneur  

def job_group(x):
    if x == 'admin.' or x == 'management' or x == 'services':
        return 'White Collar'
    elif x == 'blue-collar' or x  == 'housemaid' or x == 'technician':
        return 'Blue Collar'
    elif x == 'entrepreneur' or x == 'self-employed':
        return 'Entrepreneur'
    else:
        return x
```

```python
# applying function on job column

bank_df['Job_type'] = bank_df['Job_type'].apply(job_group)
```

```python
# Checking unique values of job column
bank_df['Job_type'].unique()
```

```python
# Checking value counts  of job column

bank_df['Job_type'].value_counts()
```

```python
# checking  null values from column

bank_df['education'].value_counts()
```

```python
# creating function for replacing null values

def unknown_secondary(x):
    if x == 'unknown':
        return 'secondary'
    else:
        return x
```

```python
# applying function on column
    
bank_df['education'] = bank_df['education'] .apply(unknown_secondary)
```

```python
# checking value counts for null values

bank_df['education'].value_counts()
```

```python
# creating function for balance group
# considered value less than 0 as negstive balance
# considerd value greater than 0 and less than equal to  500 as low balance
# considerd value greater than 500 and less than equal to  4000 as average balance
# considerd value greater than 4000  as high  balance

def balance_grp(x):
    if x <= 0 :
        return 'Negative balance'
    elif (x > 0 and x <=500):
        return 'Low Balance' 
    elif (x > 500 and x <= 4000):
        return 'Average Balance'
    else: 
        return 'High Balance'
```

```python
# applying function on column

bank_df['balance']=bank_df['balance'].apply(balance_grp)
```

```python
# Checking unique values of balance column

bank_df['balance'].unique()
```

```python
# Checking value count of balance column

bank_df['balance'].value_counts()
```

```python
# given duration is in second 
# we have converted that second in minutes and round it to zero 

bank_df['last_call_duration'] =  (bank_df['last_call_duration'] / 60).round(0)
```

```python
# droping contact column 

bank_df.drop(['contact'],axis=1 ,inplace = True)
```

```python
# creating function for balance group
# considerd duration greater than equal to 0 and less than equal to 2 as short call time
# considerd duration greater than 2 and less than equal to 5 as medium call time
# considerd duration greater than 5 as high call time
def duration(x):
    if (x >= 0 and x <= 2):
        return 'Short Call Time'
    elif (x>2 and x <=5):
        return 'Medium Call Time'
    else:
        return 'High Call Time'
```

```python
# applying function on column

bank_df['last_call_duration'] = bank_df['last_call_duration'].apply(duration) 
```

```python
# checking unique value from column 

bank_df['last_call_duration'].unique()
```

```python
# cheking value count of column 

bank_df['last_call_duration'].value_counts()
```

```python
# creating function for balance group
# cosidered value greater tha 0 and less than equal to 5 as upto 5 followups
# considered value greater than 5 as more than 5 followups

def campaign_grp(x):
    if x > 0 and x <=5:
        return 'Upto 5 Follow Ups'
    else:
        return 'More Than 5 Follow Ups'
```

```python
# applying function on column 

bank_df['Current_followUp'] = bank_df['Current_followUp'].apply(campaign_grp)
```

```python
# cheking value counts of column

bank_df['Current_followUp'].value_counts()
```

```python
# creating function for Contact_Day_diff column group

def pdays_grp(x):
    if x == -1:
        return 'Not Contacted'
    elif x >= 0 and x <= 90:
        return '0-3 Months Back'
    elif x > 90 and x <= 180:
        return '3-6 Months Back'
    else:
        return 'More Than 6 Months'
```

```python
# applying function on pdays column

bank_df['Contact_Day_diff'] = bank_df['Contact_Day_diff'].apply(pdays_grp)
```

```python
bank_df['Contact_Day_diff'].unique()
```

```python
bank_df['Contact_Day_diff'].value_counts()
```

```python
# creating function for pdays column group

def previous_grp(x):
    if x >= 0 and x <=5:
        return 'Upto 5 Follow Ups'
    else:
        return 'More Than 5 Follow Ups'
```

```python
# applying function on prevoius column
bank_df['previous_followUp'] = bank_df['previous_followUp'].apply(previous_grp)
```

```python
# repllacing fialure and sucess with some meaningfull names
bank_df['Previous_camp_Status'] = bank_df['Previous_camp_Status'].replace('failure','P N Subscribed').replace('success','P Subscribed ')
```

```python
bank_df['Previous_camp_Status'].value_counts()
```

```python
# also replaing for output column 

bank_df['Current_camp_status'] = bank_df['Current_camp_status'].replace('yes','Subscribed').replace('no','Not Subscribed ')
```
